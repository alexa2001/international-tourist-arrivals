```python
import matplotlib.pyplot as plt
import pandas as pd
import csv

data = pd.read_csv('international-tourist-arrivals-by-world-region.csv')
# data1 = pd.read_csv('technology-adoption-by-households-in-the-united-states.csv')

arrivals=[]
region=[]

with open("international-tourist-arrivals-by-world-region.csv", "r") as input_file:
    reader = csv.reader(input_file)
    next(reader) #skip the header name
    for entity, code, year, num_arrivals in reader:
        if year == '1950':
            arrivals.append(num_arrivals)
            region.append(entity)
            
plt.pie(arrivals,labels=region,autopct='%1.1f%%')
plt.title('International arrivals by region in 1950')
plt.axis('equal')
plt.show()          
        
```
